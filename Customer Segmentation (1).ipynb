# K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

max_k = 13
silhoutte_scores=[]
inertia=[]
k_values = range(2,max_k + 1)

for k in k_values:
    kmeans= KMeans(n_clusters=k, random_state=42,max_iter=1000)
    cluster_labels = kmeans.fit_predict(scaled_df)
    sil_score = silhouette_score(scaled_df,cluster_labels)
    silhoutte_scores.append(sil_score)
    inertia.append(kmeans.inertia_)


plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.plot(k_values,inertia,marker='o')
plt.xlabel('No  of Clusters')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(k_values,silhoutte_scores,marker='o')
plt.xlabel('No  of Clusters')
plt.grid(True)
```

```python
kmeans = KMeans(n_clusters=4,random_state=42, max_iter=1000)
cluster_labels=kmeans.fit_predict(scaled_df)
cluster_labels
```

```python
not_outliers_df['Cluster'] = cluster_labels
not_outliers_df
```

```python
cluster_colors = {'0':'blue','1':'orange','2':'green','3':'red'}
colors = not_outliers_df['Cluster'].map(cluster_colors)
```

```python
fig = plt.figure(figsize=(10,10))
ax=fig.add_subplot(projection='3d')

scatter = ax.scatter(not_outliers_df['MonetaryValue'],not_outliers_df['Frequency'],not_outliers_df['Recency'],c=colors,marker ='o')

ax.set_xlabel('Monetary  Value')
ax.set_ylabel('Frequency')
ax.set_zlabel('Recency')
```

# Silhouette Score (Range: -1 to 1)
Measures: How similar a data point is to its own cluster compared to other clusters.

Higher is better:

~0.7–1.0: Strong structure

~0.5–0.7: Reasonable structure

< 0.25: Weak or no structure (clusters may overlap)

Interpretation: A score of 0.61 means that your clusters are well-separated and compact, indicating good clustering quality.

# Calinski-Harabasz Index (CHI) (Higher = better)
Measures: Ratio of between-cluster dispersion to within-cluster dispersion.

Higher values indicate:

Clusters are dense and well-separated.

No strict range, but it's used to compare between different K values.

Interpretation: A CHI score of 2900+ (like yours) is excellent, especially if it's much higher than other K values.

# Davies–Bouldin Index (Lower = better)
Measures: Average similarity of each cluster with the most similar other cluster.

Values:

0.3–0.6 → Good

< 1.0 → Poor cluster separation

Interpretation: A score like 0.4–0.6 shows that clusters are distinct and well-formed. If you’re above 1.0, clusters are overlapping too much.

```python
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score

# Fit KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(scaled_df)
scaled_df['Cluster'] = labels
X_scaled = scaled_df
# Metrics
silhouette = silhouette_score(X_scaled, labels)
calinski = calinski_harabasz_score(X_scaled, labels)
davies = davies_bouldin_score(X_scaled, labels)

print(f"Silhouette Score: {silhouette:.3f}")
print(f"Calinski-Harabasz Index: {calinski:.3f}")
print(f"Davies-Bouldin Index: {davies:.3f}")
```

A violin plot is a hybrid of:

a box plot (showing median, interquartile range, etc.)

and a kernel density estimate (KDE) on each side, showing the distribution (shape) of the data.

It gives a richer view of data distribution compared to box plots alone.

🧠 Why Use Violin Plots in Clustering Analysis?
To compare the distribution of features like Monetary, Frequency, and Recency across clusters.

Helps you understand behavioral differences between customer groups.

For example, you might see that Cluster 0 has high monetary but low frequency, suggesting big-spending occasional customers, while Cluster 2 may represent frequent low spenders.

📚 Breakdown of Each Plot
1. MonetaryValue by Cluster
X-axis: Cluster labels (e.g., 0, 1, 2...)

Y-axis: Monetary Value (spending)

Helps identify which clusters contain high-value vs low-value customers.

2. Frequency by Cluster
X-axis: Cluster

Y-axis: Number of purchases

Shows which clusters contain frequent shoppers vs infrequent ones.

3. Recency by Cluster
X-axis: Cluster

Y-axis: Recency

Lower recency = more recent purchase → fresh engagement.

Helps to identify active vs dormant customers per cluster.

🧪 How to Interpret the Plots
Width of the violin: How many data points fall in that range.

Long and narrow tails: Presence of outliers or sparse data.

Tall and fat violins: A broad spread of values in that cluster.

Position of median line: Central tendency (like a boxplot).



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 18))

# Monetary Value
plt.subplot(3, 1, 1)
sns.violinplot(x='Cluster', y='MonetaryValue', data=not_outliers_df, palette=cluster_colors)
plt.title('Monetary Value by Cluster')
plt.ylabel('Monetary Value')

# Frequency
plt.subplot(3, 1, 2)
sns.violinplot(x='Cluster', y='Frequency', data=not_outliers_df, palette=cluster_colors)
plt.title('Frequency by Cluster')
plt.ylabel('Frequency')

# Recency
plt.subplot(3, 1, 3)
sns.violinplot(x='Cluster', y='Recency', data=not_outliers_df, palette=cluster_colors)
plt.title('Recency by Cluster')
plt.ylabel('Recency')

plt.tight_layout()
plt.show()
```
